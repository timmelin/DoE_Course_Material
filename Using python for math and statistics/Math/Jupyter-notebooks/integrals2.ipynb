## New Tricks for Substitutions

When performing more complicated integration by parts, there are a few tips and tricks.

### To help pick the "u" for integration, there's a priority list ["LIATE"](https://en.wikipedia.org/wiki/Integration_by_parts#LIATE_rule):
1. Logarithm
2. Inverse trigonometric
3. Algebraic
4. Trigonometric
5. Exponential

In other words, pick a "u" term in the list, and the "dv" comes later in the list.

### Repeated Integration by Parts via Table

Sometimes you need to do repeated integration by parts because the new integral also requires integration by parts:

$$
\int u d v=u v-\int v d u
$$

but the $\int v d u$ part also requires integration by parts...

For example:

$$
\int x^{3} \cos x d x
$$

The LIATE rules suggest we want to take the $x^3$ as $u$ and the $\cos x dx$ as $dv$.

Then we set up a little table - we have one column with $u$ and its derivatives and another column with the integrals of $dv$:

$$
\begin{array}{|c|c|c|c|}
\hline \# & \text { Sign } & \text { A: derivatives } u^{(i)} & \text { B: integrals } v^{(n-1)} \\
\hline 0 & + & x^{3} & \cos x \\
\hline 1 & - & 3 x^{2} & \sin x \\
\hline 2 & + & 6 x & -\cos x \\
\hline 3 & - & 6 & -\sin x \\
\hline 4 & + & 0 & \cos x \\
\hline
\end{array}
$$

Then you connect each line, matching up $u$ and it's derivatives with the corresponding $v$ and its integrals:

<img src="./images/tabular-integration.png" width="500" />

So this particular integral:

$$
\int x^{3} \cos x d x
$$

Evaluates to:

$$
x^3 \sin x + 3x^2 \cos x - 6x \sin x -6 \cos x + C
$$

You can finish the table three ways:
1. You get zero out of the derivative (so any remaining terms will be zero)
2. You get a row that you can integrate trivially
3. You get a row that is some multiple of the original row

As an example of the second option, consider:

$$
\int \ln x dx
$$

We should pick $\ln x$ from the "LIATE" rules:

$$
\begin{array}{|c|c|c|c|}
\hline \# & \text { Sign } & \text { A: derivatives } u^{(i)} & \text { B: integrals } v^{(n-1)} \\
\hline 0 & + & \ln x & 1 \\
\hline 1 & - & 1/x & x \\
\hline
\end{array}
$$

While we can't continue to take derivatives of $1/x$ to get a zero, we can see that:

$$
x \ln x - \int \frac{x}{x} dx
$$

is simply

$$
x \ln x - x + C
$$

Finally, if you have both exponentials and trigonometric functions:

$$
\int e^{x} \cos x d x
$$

It doesn't really matter if you pick $e^x$ or $\cos x$ as u:

$$
\begin{array}{|c|c|c|c|}
\hline \# i & \text { Sign } & \text { A: derivatives } u^{(i)} & \text { B: integrals } v^{(n-1)} \\
\hline 0 & + & e^{x} & \cos x \\
\hline 1 & - & e^{x} & \sin x \\
\hline 2 & + & e^{x} & -\cos x \\
\hline
\end{array}
$$

Notice that the final row is simply a multiple of the original integral.

$$
\underbrace{\int e^{x} \cos x d x}_{\text {step } 0}=\underbrace{(+1)\left(e^{x}\right)(\sin x)}_{i=0}+\underbrace{(-1)\left(e^{x}\right)(-\cos x)}_{i=1}+\underbrace{\int(+1)\left(e^{x}\right)(-\cos x) d x}_{i=2}
$$

then:

$$
2 \int e^{x} \cos x d x=e^{x} \sin x+e^{x} \cos x+C
$$

### Integration over Multiple Variables

We live in a 3D world, and often need to integrate over more than one variable. For example:

$$
\int_{c}^{d} \int_{a}^{b} f(x, y) d x d y=\int_{c}^{d}\left[\int_{a}^{b} f(x, y) d x\right] d y
$$

Much like performing partial derivatives, this just says:
- integrate with respect to x, considering all other variables as constants
- then integrate *that* result with respect to y, considering all other variables as constants

Generally, we work with functions for which the order of integration does not matter (much like it didn't matter if we took partial derivatives w.r.t. x and then y or y and then x).

Here's an example - find the area between the two curves:

<img src="./images/double-integration.png" width="300" />

$$
\int_{0}^{4} \int_{y / 2}^{\sqrt{y}} d x d y=\int_{0}^{4}\left[\int_{y / 2}^{\sqrt{y}} d x\right] d y
$$

then

$$
\int_{0}^{4}\left[\int_{y / 2}^{\sqrt{y}} d x\right] d y=\int_{0}^{4}[x]_{y / 2}^{\sqrt{y}} d y=\int_{0}^{4}\left(\sqrt{y}-\frac{y}{2}\right) d y
$$

It works out to be 4/3 - either if you evaluate x first then y, or y first and then x.

$$
\int_{0}^{2} \int_{x^{2}}^{2 x} d y d x=\int_{0}^{2}\left[\int_{x^{2}}^{2 x} d y\right] d x=\int_{0}^{2}\left(2 x-x^{2}\right) d x=\frac{4}{3}
$$

### Choice of Variables / Coordinate Systems

While evaluating double integrals in Cartesian coordinates is relatively straight-forward, depending on the function, we may find it easier in polar coordinates.

<img src="./images/area-circle.png" width="300">

To evaluate this in Cartesian space, we need:

$$
\int_{0}^{1} \int_{0}^{\sqrt{1-y^{2}}} d x d y=\int_{0}^{1} \sqrt{1-y^{2}} d y
$$

I certainly don't know how to evaluate that integral. It turns out, it's known - you can find it via Sympy or a table of integrals.

$$
\int_{0}^{1} \sqrt{1^{2}-x^{2}} d x=\frac{1}{2} \sin ^{-1}(1)=\frac{1}{4} \pi
$$

But the integral (and limits) would be much easier in polar 2D coordinates:
- $0 < r < 1$
- $0 < \theta < \pi/2$

$$
\int_{0}^{1} \int_{0}^{\sqrt{1-y^{2}}} d x d y=\int_{0}^{\pi / 2} \int_{0}^{1} r d r d \theta
$$

<div class="alert alert-block alert-success">

The key with integrals in non-Cartesian coordinates is to remember that the area or volume element isn't just $dx dy$:
    
- 2D Polar Coordinates: $f(r, \theta) r dr d\theta$
- 3D Spherical Coordinates: $f(r, \theta, \varphi) r^{2} \sin \theta dr d\theta d\varphi$

</div>

For example, the volume of a sphere:

$$
V=\int_{0}^{2 \pi} \int_{0}^{\pi} \int_{0}^{a} r^{2} \sin \theta d r d \theta d \varphi=\int_{0}^{2 \pi} d \varphi \times \int_{0}^{\pi} \sin \theta d \theta \times \int_{0}^{a} r^{2} d r
$$


I can't say this clearly enough.

This is the ***most frequent mistake*** in physical chemistry.

Integrate $e^{-\alpha r}$ in spherical coordinates:

$$
\int_{0}^{2 \pi} \int_{0}^{\pi} \int_{0}^{a} e^{-\alpha r} r^{2} \sin \theta d r d \theta d \varphi
$$

Notice that the $r^{2} \sin \theta d r d \theta d \varphi$ is the volume element - it's the replacement for $dx dy dz$. We need this each and *every* time we integrate in spherical coordinates.

One more time...

The function we want to integrate is $e^{-\alpha r}$ and the $r^{2} \sin \theta d r d \theta d \varphi$ comes along with spherical coordinates.

```python
from sympy import init_session
init_session()
```

The main command for `sympy` is `integrate(function, variable)`

This can also work as:
`integrate(function, (variable, lower, upper)`

And for multiple integration, it's similar to `diff()`:

`integrate(function, (variable, lower, upper), (variable2, lower, upper)`

```python
# integrate the area of a circle of size 1
r, theta = symbols('r theta')
f = 1
integrate(f * r, (r, 0, 1), (theta, 0, 2*pi))
```

```python
alpha, r, theta, phi = symbols('alpha r theta phi')
f = exp(-2*r)

integrate(f * r**2 * sin(theta), (r, 0, 1), (theta, 0, pi), (phi, 0, 2*pi))
# notice that we have the function times r**2 times sin(theta) anytime we do spherical int
```

## Take-home message:

Multiple integration isn't bad. It's much like partial derivatives - you do one variable, then you do the next.

The annoying part of 2D and 3D integrals comes when you perform them for polar 2D and spherical 3D - remember the limits of the integrals and that $dx dy$ in 2D and $dx dy dz$ are replaced:

- 2D Polar Coordinates: $f(r, \theta) r dr d\theta$
- 3D Spherical Coordinates: $f(r, \theta, \varphi) r^{2} \sin \theta dr d\theta d\varphi$
